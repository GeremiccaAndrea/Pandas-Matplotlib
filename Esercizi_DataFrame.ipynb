```python
import pandas as pd 
```

```python
df = pd.read_json('https://raw.githubusercontent.com/wtitze/4E/main/comuni_italiani.json')
df
```

```python
#numero di comuni italiani 
df.count()
```

```python
#il numero di province italiane 
len(set(df['Provincia'])) #con il set rimuovo i duplicati 
```

```python
#numero di province per ogni regione
dfregione = df.groupby(['Denominazione Regione' , 'Provincia']).count()[['Ripartizione geografica']]
dfregione.groupby('Denominazione Regione').count()
```

```python
#il numero di comuni per provincia, ordinarli in ordine decrescente
```

```python
#la provincia con meno comuni
dfregione[dfregione['Denominazione in italiano'] == dfregione['Denominazione in italiano'].min()[['Denominazione in italiano']]]
```

```python
#il numero medio di comuni nelle varie province 
dfregione.mean()
```

```python
#grafico a barre orizzontali dell'esercizio 3 
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels =  dfregione.index
dati = dfregione['Ripartizione geografica']
fig, ax = plt.subplots(figsize=(15,8))
ax.barh(labels, dati)
```
