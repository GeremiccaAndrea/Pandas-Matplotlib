```python
#il numero di casi totali in Lombardia
df[df['denominazione_regione'] == 'Lombardia'].sum()[['totale_casi']]
```

```python
#il numero di casi totali in ogni regione. Ordinare in ordine decrescente in base al numero di casi
dfgroup = df.groupby('denominazione_regione')[['totale_casi']].sum().sort_values(by= 'totale_casi', ascending = False)
dfgroup
```

```python
#le regioni che hanno più di un milione di casi
dfgroup[dfgroup['totale_casi'] > 1000000]
```

```python
#il numero di casi in Italia
dfgroup['totale_casi'].sum()
```

```python
#la percentuale di casi in ogni regione rispetto al totale della popolazione italiana
dfgroup/ dfgroup['totale_casi'].sum() *100
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/consegne-vaccini-latest.csv ')
df
```

```python
#visualizzare il numero totale di record presenti nel file
df.count()
```

```python
#visualizzare il numero di record presenti relativi ad una regione inserita dall'utente
Nome_Regione = input('Inserire il nome di una regione: ')
df[df['reg'] == Nome_Regione ].count()
```

```python
#modificare l'esercizio precedente per rispondere con un messaggio appropriato se la regione non è presente
Nome_Regione = input('Inserire il nome di una regione: ')
risultato= df[df['reg'] == Nome_Regione ].count()['area']
if risultato == 0: 
  print('La regione non è presente')
else:
  print(risultato)
```

```python
#visualizzare il numero totale di vaccini consegnati in una regione inserita dall'utente
Nome_Regione = input('Inserire il nome di una regione: ')
risultato= df[df['reg'] == Nome_Regione ].sum()['numero_dosi']
risultato
```

```python
#visualizzare il numero totale di vaccini per ogni fornitore
dfgb = df.groupby('forn')[['numero_dosi']].sum()
dfgb
```

```python
#visualizzare il numero di vaccini per ogni fornitore di una regione inserita dall'utente. Ordinare in ordine decrescente sul numero di vaccini
Nome_Regione = input('Inserire il nome di una regione: ')
dfreg = df[df['reg'] == Nome_Regione]
dfreg.groupby('forn')[['numero_dosi']].sum()
```
